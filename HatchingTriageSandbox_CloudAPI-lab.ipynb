1. get sample list

**Example: the last 10 public samples index from tria.ge**

```python
''' 1. get sample list '''
import pandas as pd
import requests
from requests.packages.urllib3.exceptions import InsecureRequestWarning
requests.packages.urllib3.disable_warnings(InsecureRequestWarning)
 
baseUrl = "https://tria.ge/api/v0/"
url = baseUrl + "samples?&subset=public&limit=10&offset=0"

headers = { 
    'Authorization' : "Bearer " + ak
}

r = requests.get(url, headers=headers,verify=False)
jsonResponse = r.json()
df = pd.json_normalize(jsonResponse['data'])
df
```

3. get config by sample


"dropper" and "config" keys are stored under extracted key which value is json array.


```python
''' 3. get config by sample '''

import json
import requests
from requests.packages.urllib3.exceptions import InsecureRequestWarning
requests.packages.urllib3.disable_warnings(InsecureRequestWarning)
 
baseUrl = "https://tria.ge/api/v0/"
url = baseUrl + "samples/230317-1nw93scb6x/overview.json"

headers = { 
    'Authorization' : "Bearer " + ak
}

r = requests.get(url, headers=headers,verify=False)
jsonResponse = r.json()
for extracted in jsonResponse["extracted"]:
    if "config" in extracted: 
        config = extracted["config"]
        print(json.dumps(config, indent=2))
```

```python
''' 4. search report (family:emotet)'''

import pandas as pd
import requests
from requests.packages.urllib3.exceptions import InsecureRequestWarning
requests.packages.urllib3.disable_warnings(InsecureRequestWarning)
 

baseUrl = "https://tria.ge/api/v0/"
url = baseUrl + "search?query=family:emotet"

headers = { 
    'Authorization' : "Bearer " + ak
}

r = requests.get(url, headers=headers,verify=False)
jsonResponse = r.json()
df = pd.json_normalize(jsonResponse['data'])
df
```
